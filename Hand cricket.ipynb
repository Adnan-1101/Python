### Hand Cricket

```python
import random

print("\033[1mWelcome to Hand cricket\033[0m")

print('''The rules of the game are:
      1. There will be a toss between you and the computer, you have to chose either odd or even, after that you should put a number between 1 to 10
    if the sum of your and the computer's number is odd or even, accordingly to what you have choosen you will either win or loss the toss.
    If you win the toss you can either bat or bowl, and the same thing will happen with the computer.
    2. If you are batting then you have to put a number between 1 to 10 to score runs, if the number you have put and the number computer has put is same
    the you will be declared out, if not then you can continue to bat till you get out and it is a 10 over game
    3. After you finish batting, you have to bowl and defend your score and get the computer out be chosing the same number as the computer
    4. This is the beginning of your successful cricket journey
    5.Hope you have fun''')



player_name = input("Enter your name: ")

while True:
    player = input("Enter Odd or Even: ").lower()

    if player == "odd" or player == "even":
        break
    else:
        print("\033[1mInvalid Input!!\033[0m\nPlease enter Odd or Even only\n")

while True:
    choose = input("Enter a number from 1-6!: ")
    if choose.isdigit() and 1 <= int(choose) <= 6:
        choose = int(choose)
        break
    else:
        print("\033[1mInvalid Input!!\033[0m\nPlease enter numbers from 1-6 only\n")

sys = random.randint(1, 6)
print(f"You Entered: {choose}\nComputer has chosen: {sys}")
add = choose + sys
choice = ["Batting", "Balling"]
sys_choice = random.choice(choice)

def batting():
    global GameOver
    GameOver = False
    print("It's your batting!!")
    print("Game started")
    player_score = 0
    computer_runs = 0
    while True:
        player = input("Enter the number in between 1-6!")
        if player.isdigit() and 1 <= int(player) <= 6:
            player = int(player)
        else:
            print("\033[1mInvalid Input!!\033[0m\nPlease enter numbers from 1-6 only\n")
            continue
        system = random.randint(1, 6)
        print(f"You entered: {player}\nComputer chosen: {system}")
        player_score += player
        print("Your current runs:", player_score)
        print("---------------------------------------------------------------------------------")
        if player == system:
            print("OOPS!\nYou are out, Computer has taken your wicket")
            print(f"Computer is ready for batting\nComputer needs {player_score + 1} to win!")
            print("--------------------------------------------------------------------------------------------")
            print("--------------------------------------------------------------------------------------------------------")
            break
    while True:
        player = input("Enter a number in between 1-6 for bowling")
        if player.isdigit() and 1 <= int(player) <= 6:
            player = int(player)
        else:
            print("\033[1mInvalid Input!!\033[0m\nPlease enter numbers from 1-6 only\n")
            continue
        system = random.randint(1, 6)
        print(f"You entered: {player}\nComputer chosen: {system}")
        computer_runs += system
        print("Computer current runs:", computer_runs)
        print("--------------------------------------------------------------------------------------------------------")
        if computer_runs > player:
            print(f"You loss!\n\033[1mComputer has won the match by {computer_runs - player_score} runs!!\033[0m ")
            print('-------------------------------------------------------------------------------------------')
            print('--------------------------------------* \033[1mMatch Ended\033[0m *--------------------------------------')
            print('-------------------------------------------------------------------------------------------')
            GameOver = True
            break
        elif player == system:
            print(f"Hurray!!\n{player_name} won the match by {player_score - computer_runs} runs")
            print('-------------------------------------------------------------------------------------------')
            print('--------------------------------------* \033[1mMatch Ended\033[0m *--------------------------------------')
            print('-------------------------------------------------------------------------------------------')
            GameOver = True
            break

def balling():
    global GameOver
    GameOver = False
    print("Be prepared for balling!!")
    print("_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_> \033[1mLet's begin!!\033[0m <-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-\n")
    player_score = 0
    system_runs = 0
    while True:
        player = input("Enter a number between 1-6 for balling!!")
        if player.isdigit() and 1 <= int(player) <= 6:
            player = int(player)
        else:
            print("\033[1mInvalid Input!!\033[0m\nPlease enter numbers from 1-6 only\n")
            continue
        system = random.randint(1, 6)
        print(f"You entered {player}\nComputer chosen {system}")
        system_runs += system
        print("Computer runs:", system_runs)
        print('-------------------------------------------------------------------------------------------')
        if player == system:
            print("Great!!,You got computer wicket\nComputer are done with batting")
            print(f"Are you ready for batting\nYou needs {system_runs + 1} to win!")
            print('-------------------------------------------------------------------------------------------')
            print('-------------------------------------------------------------------------------------------')
            break
    while True:
        player = input("Enter the number in between 1-6 for batting")
        if player.isdigit() and 1 <= int(player) <= 6:
            player = int(player)
        else:
            print("\033[1mInvalid Input!!\033[0m\nPlease enter a valid input!!\n")
            continue
        system = random.randint(1, 6)
        print(f"You Entered {player}\n Computer chosen {system}")
        player_score += player
        print("Your current runs:", player_score)
        print("-------------------------------------------------------------------------------------------------")
        if player_score > system_runs:
            print(f"Hurray!!\n{player_name} won the match by {player_score - system_runs} runs")
            print('-------------------------------------------------------------------------------------------')
            print('--------------------------------------* \033[1mMatch Ended\033[0m *--------------------------------------')
            print('-------------------------------------------------------------------------------------------')
            GameOver = True
            break
        elif player == system:
            print("Computer has taken your wicket!!")
            print(f"\033[1mComputer won the match by {system_runs - player_score} runs!!\033[0m")
            print("Better luck next time!!")
            print('-------------------------------------------------------------------------------------------')
            print('--------------------------------------* \033[1mMatch Ended\033[0m *--------------------------------------')
            print('-------------------------------------------------------------------------------------------')
            GameOver = True
            break

if add % 2 == 0:
    if player == "even":
        BatOrBall = input("\033[1mYou won the toss!!!\033[0m\nYou have the choice to choose Batting/Bowling\n1 for Batting\n2 for Bowling\n ")
        if BatOrBall == "1":
            batting()
        elif BatOrBall == "2":
            balling()
        else:
            print("\033[1mInvalid Input!!\033[0m\nPlease enter a valid input!!")
    else:
        print(f"Computer has won the toss\nComputer chosen {sys_choice}")
        if sys_choice == "Batting":
            balling()
        else:
            batting()
else:
    if player == "odd":
        BatOrBall = input("\033[1mYou won the toss!!!\033[0m\nYou have the choice to choose Batting/Bowling\n1 for Batting\n2 for Bowling\n ")
        if BatOrBall == "1":
            batting()
        elif BatOrBall == "2":
            balling()
        else:
            print("\033[1mInvalid Input!!\033[0m\nPlease enter a valid input!!")
    else:
        print(f"Computer has won the toss\nComputer chosen {sys_choice}")
        if sys_choice == "Batting":
            balling()
        else:
            batting()
```
